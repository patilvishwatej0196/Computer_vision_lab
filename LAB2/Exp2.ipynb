```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('car.jpg')
plt.imshow(image)
plt.show()


```

```python
height, width,  = image.shape

for i in range(0, height - 1):
    for j in range(0, width - 1):

        # Get the pixel value
        pixel = image[i, j]

        # Negate each channel by
        # subtracting it from 255

        # 1st index contains red pixel
        pixel[0] = 255 - pixel[0]

        # 2nd index contains green pixel
        pixel[1] = 255 - pixel[1]

        # 3rd index contains blue pixel
        pixel[2] = 255 - pixel[2]

        # Store new values in the pixel
        image[i, j] = pixel
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


image = cv2.imread('car.jpg')

# Find the minimum and maximum pixel values in the image
R_min = np.min(image)
R_max = np.max(image)

# Calculate alpha (scale) and beta (bias/offset)
alpha = 255.0 / (R_max - R_min) if (R_max - R_min) != 0 else 0
beta = -alpha * R_min

# Apply the transformation g(i, j) = alpha*f(i, j) + beta using cv2.convertScaleAbs
# This function handles the saturation of values to the [0, 255] range
stretched_image_manual = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Display the result
cv2_imshow(image)
cv2_imshow(stretched_image_manual)
```
